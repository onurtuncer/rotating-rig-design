## Physics Background

### Euler Turbomachinery Equation

The fundamental equation relating shaft work to the change in angular momentum:

$$
W = U \cdot \Delta C_\theta = C_p \cdot \Delta T_0
$$

| Symbol | Description | Unit |
|---|---|---|
| \( W \) | Specific work input | J/kg |
| \( U \) | Blade (peripheral) speed at mean radius | m/s |
| \( \Delta C_\theta \) | Change in whirl (swirl) velocity | m/s |
| \( C_p \) | Specific heat at constant pressure | J/(kg·K) |
| \( \Delta T_0 \) | Total temperature rise across rotor | K |

---

### Pressure Ratio — Isentropic Relation

$$
PR = \frac{P_{02}}{P_{01}} = \left(1 + \eta_{is} \cdot \frac{\Delta T_0}{T_{0,in}}\right)^{\frac{\gamma}{\gamma - 1}}
$$

Rearranged to give total temperature rise from a target PR:

$$
\Delta T_{0,is} = T_{0,in} \cdot \left(PR^{\frac{\gamma - 1}{\gamma}} - 1\right)
$$

$$
\Delta T_0 = \frac{\Delta T_{0,is}}{\eta_{is}}
$$

---

### Dimensionless Coefficients

| Coefficient | Formula | Typical Range | Notes |
|---|---|---|---|
| Flow coefficient | \( \phi = \frac{C_a}{U} \) | 0.35 – 0.65 | Too low → stall risk; too high → high Mach |
| Work (loading) coefficient | \( \psi = \frac{W}{U^2} \) | 0.20 – 0.50 | High \( \psi \) → high blade loading |
| De Haller number | \( DH = \frac{W_2}{W_1} \) | > 0.72 | Minimum to avoid diffusion stall |
| Lieblein diffusion factor | \( DF = 1 - \frac{W_2}{W_1} + \frac{\Delta C_\theta}{2\sigma W_1} \) | < 0.45 | Requires solidity \( \sigma \) |

---

### Velocity Triangles (assuming \( C_{\theta 1} = 0 \), i.e., swirl-free inlet)

$$
\beta_1 = \arctan\left(\frac{U}{C_a}\right)
$$

$$
\beta_2 = \arctan\left(\frac{U - C_{\theta 2}}{C_a}\right)
$$

$$
\alpha_2 = \arctan\left(\frac{C_{\theta 2}}{C_a}\right)
$$

$$
\Delta \beta = \beta_1 - \beta_2
$$

Where:

$$
C_{\theta 2} = \frac{W}{U}
$$

---

### Annulus Geometry

$$
r_{mean} = \frac{1}{2}(r_{tip} + r_{hub})
$$

$$
h = r_{tip} - r_{hub}
$$

$$
\nu = \frac{r_{hub}}{r_{tip}}
$$

$$
A = \pi \left(r_{tip}^2 - r_{hub}^2\right)
$$

$$
\dot{m} = \rho \cdot C_a \cdot A
$$

$$
P_{shaft} = \dot{m} \cdot W
$$

---

### Free Vortex Law (Radial Distribution)

The simplest and most common radial design law. Whirl velocity varies as:

$$
r \cdot C_\theta = \text{constant}
$$

$$
C_{\theta 2}(r) = C_{\theta 2,mean} \cdot \frac{r_{mean}}{r}
$$

This ensures constant specific work at every radius and zero radial pressure gradient in the axial direction.

## Design Workflow

## Design Workflow

### 1. Input Parameters
- Tip diameter: \( D_{tip} \)
- Rotational speed: \( N \)
- Target pressure ratio: \( PR \)
- Isentropic efficiency: \( \eta_{is} \)
- Hub-to-tip ratio: \( \nu \)

---

### 2. Annulus Geometry
Compute geometric quantities:

$$
r_{tip} = \frac{D_{tip}}{2}
$$

$$
r_{hub} = \nu \cdot r_{tip}
$$

$$
r_{mean} = \frac{1}{2}(r_{tip} + r_{hub})
$$

$$
h = r_{tip} - r_{hub}
$$

$$
A = \pi (r_{tip}^2 - r_{hub}^2)
$$

---

### 3. Thermodynamics (Euler Work)
From pressure ratio:

$$
\Delta T_{0,is} = T_{0,in} \left(PR^{\frac{\gamma - 1}{\gamma}} - 1\right)
$$

$$
\Delta T_0 = \frac{\Delta T_{0,is}}{\eta_{is}}
$$

$$
W = C_p \cdot \Delta T_0
$$

Whirl velocity:

$$
C_{\theta 2} = \frac{W}{U}
$$

---

### 4. Dimensionless Parameters

$$
\phi = \frac{C_a}{U}, \quad
\psi = \frac{W}{U^2}
$$

---

### 5. Performance Checks
- De Haller:
  $$
  DH = \frac{W_2}{W_1} > 0.72
  $$
- Mach number limits
- Blade height constraint:
  $$
  h \geq h_{min}
  $$

---

### 6. Mass Flow and Shaft Power

$$
\dot{m} = \rho \cdot C_a \cdot A
$$

$$
P_{shaft} = \dot{m} \cdot W
$$

---

### 7. Velocity Triangles (Meanline)

$$
\beta_1 = \arctan\left(\frac{U}{C_a}\right)
$$

$$
\beta_2 = \arctan\left(\frac{U - C_{\theta 2}}{C_a}\right)
$$

$$
\alpha_2 = \arctan\left(\frac{C_{\theta 2}}{C_a}\right)
$$

$$
\Delta \beta = \beta_1 - \beta_2
$$

---

### 8. Radial Distribution (Free Vortex)

$$
C_{\theta 2}(r) = C_{\theta 2,mean} \cdot \frac{r_{mean}}{r}
$$

---

### 9. Motor Selection
- Required shaft power: \( P_{shaft} \)
- Select motor class (direct drive / geared)

## Design Constraints Summary

| Constraint | Limit | Rationale |
|------------|-------|-----------|
| Blade height     | h ≥ 50 mm   | Mechanical and aerodynamic viability |
| De Haller number | DH ≥ 0.72   | Avoid separation / diffusion stall |
| Tip Mach number  | M_tip < 0.60 | Subsonic relative flow (no shock losses) |
| Flow coefficient | 0.35 ≤ φ ≤ 0.65 | High-efficiency operating region |
| Work coefficient | 0.15 ≤ ψ ≤ 0.55 | Smith chart optimal zone |
| Lieblein diffusion factor | DF < 0.45 | Acceptable blade loading (solidity dependent) |

## Bibliography

### Textbooks

1. **Dixon, S. L., & Hall, C. A.** (2014). *Fluid Mechanics and Thermodynamics of Turbomachinery* (7th ed.). Butterworth-Heinemann.
   — Primary reference for meanline analysis, velocity triangles, and stage design methodology.

2. **Cumpsty, N. A.** (2004). *Compressor Aerodynamics*. Krieger Publishing.
   — Comprehensive treatment of axial compressor aerodynamics, blade-to-blade flow, and radial equilibrium.

3. **Aungier, R. H.** (2003). *Axial-Flow Compressors: A Strategy for Aerodynamic Design and Analysis*. ASME Press.
   — Detailed design methodology including loss models and off-design performance.

4. **Cumpsty, N. A., & Heyes, A.** (2015). *Jet Propulsion* (3rd ed.). Cambridge University Press.
   — Broader propulsion context; useful for understanding stage matching and overall cycle.

5. **Lakshminarayana, B.** (1996). *Fluid Dynamics and Heat Transfer of Turbomachinery*. Wiley-Interscience.
   — Advanced treatment of 3D flow effects, secondary flows, and tip clearance losses.

6. **Çengel, Y. A., & Boles, M. A.** (2018). *Thermodynamics: An Engineering Approach* (9th ed.). McGraw-Hill.
   — Thermodynamic foundations: isentropic relations, compressor work, and efficiency definitions.

### Key Journal Papers & Standards

7. **Lieblein, S., Schwenk, F. C., & Broderick, R. L.** (1953). *Diffusion Factor for Estimating Losses and Limiting Blade Loadings in Axial-Flow Compressor Blade Elements*. NACA RM E53D01. NASA Technical Report.
   — Original derivation of the Lieblein diffusion factor; fundamental to blade loading assessment.

8. **Smith, L. H.** (1958). *The Relation Between Discharge and Losses in Diffusing Annuli*. ASME Journal of Basic Engineering.
   — Foundation of the Smith chart relating φ and ψ to stage efficiency.

9. **Carter, A. D. S.** (1950). *The Low Speed Performance of Related Aerofoils in Cascade*. Aeronautical Research Council, R&M 2865.
   — Carter's rule for blade deviation angle; still widely used in preliminary design.

10. **Howell, A. R.** (1945). *Fluid Dynamics of Axial Compressors*. Proceedings of the Institution of Mechanical Engineers, 153(1), 441–452.
    — Classic work on cascade aerodynamics and the concept of nominal conditions.

11. **De Haller, P.** (1953). *Das Verhalten von Tragflügelgittern in Axialverdichtern und im Windkanal*. Brennstoff-Wärme-Kraft, 5(10), 333–337.
    — Original paper introducing the W₂/W₁ diffusion criterion (De Haller number).

### Standards & Data Sources

12. **ICAO** (2018). *Manual of the ICAO Standard Atmosphere* (3rd ed.). International Civil Aviation Organization, Doc 7488.
    — ISA standard atmosphere definitions used for inlet conditions (T₀ = 288.15 K, P₀ = 101 325 Pa).

13. **ASME PTC 10** (1997). *Performance Test Code on Compressors and Exhausters*. American Society of Mechanical Engineers.
    — Test rig instrumentation and performance measurement standards relevant to experimental validation.
